```python
import yfinance as yf
```

```python
acao = yf.Ticker('BRAP4.SA')
```

```python
df = acao.history(start = '2021-12-01', end = '2022-12-31')
df
```

```python
df['Ano'] = df.index.year
df
```

```python
df[df.Dividends != 0]
```

```python
# Soma dos pagamentos do periodo

df.groupby(['Ano'])['Dividends'].sum()
```

```python
# Quantidade de pagamento no periodo

pagamento_dividendos = df[df.Dividends != 0]
pagamento_dividendos.groupby(['Ano'])['Dividends'].count()
```

```python
# Preço atual do ativo

ultima_cotacao = df['Close'][-1]
ultima_cotacao
```

```python
# Soma dos dividendos do periodo

dados2 = df[df.Dividends != 0]
total_dividendos = sum(dados2.Dividends)
total_dividendos
```

```python
# DIVIDEND YELD Ultimos 12 meses

(total_dividendos/ultima_cotacao)*100
```
